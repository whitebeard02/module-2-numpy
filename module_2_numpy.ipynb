```python
import numpy as np
```

```python
marks = np.random.randint(0, 101, size=(100, 5))

subjects = ["Math", "Physics", "Chemistry", "Biology", "English"]
student_ids = np.arange(1, 101)
```

```python
print("=" * 55)
print("           STUDENT ANALYTICS")
print("=" * 55)

#total marks per student
total_marks = np.sum(marks, axis=1)

#average marks per student
avg_marks = np.mean(marks, axis=1)

#top scorer
top_scorer_index = np.argmax(total_marks)
print(f"Top Scorer    : Student {student_ids[top_scorer_index]}"
      f" | Total: {total_marks[top_scorer_index]}"
      f" | Average: {avg_marks[top_scorer_index]:.2f}")

#lowest scorer
low_scorer_index = np.argmin(total_marks)
print(f"Lowest Scorer : Student {student_ids[low_scorer_index]}"
      f" | Total: {total_marks[low_scorer_index]}"
      f" | Average: {avg_marks[low_scorer_index]:.2f}")
```

```python
print("\n" + "=" * 55)
print("           SUBJECT ANALYTICS")
print("=" * 55)

#subject-wise avgs, highest, lowest
subject_avg     = np.mean(marks, axis=0)
subject_highest = np.max(marks, axis=0)
subject_lowest  = np.min(marks, axis=0)

print(f"{'Subject':<12} {'Average':>10} {'Highest':>10} {'Lowest':>10}")
print("-" * 45)
for i in range(5):
    print(f"{subjects[i]:<12}"
          f"{subject_avg[i]:>10.2f}"
          f"{subject_highest[i]:>10}"
          f"{subject_lowest[i]:>10}")
```

```python
print("\n" + "=" * 55)
print("           BOOLEAN MASKING")
print("=" * 55)

#students scoring above 80 avg
above_80_mask = avg_marks > 80
above_80_students = student_ids[above_80_mask]
print(f"\nStudents with Average Above 80 "
      f"({len(above_80_students)} students):")
print(above_80_students)

#students avg<35
failing_mask = avg_marks < 35
failing_students = student_ids[failing_mask]
print(f"\nFailing Students - Average Below 35 "
      f"({len(failing_students)} students):")
if len(failing_students) == 0:
    print("  No students failing.")
else:
    print(failing_students)

#students scoring above 90 in any subj
above_90_any_mask = np.any(marks > 90, axis=1)
above_90_students = student_ids[above_90_any_mask]
print(f"\nStudents Scoring Above 90 in At Least One Subject "
      f"({len(above_90_students)} students):")
print(above_90_students)
```

```python
print("\n" + "=" * 55)
print("           RANKING SYSTEM")
print("=" * 55)


ranked_indices = np.argsort(total_marks)[::-1]

print("\nTop 3 Students Based on Total Marks:")
print("-" * 45)
for rank in range(3):
    idx = ranked_indices[rank]
    print(f"  Rank {rank + 1} --> Student {student_ids[idx]:>3}"
          f" | Total: {total_marks[idx]}"
          f" | Average: {avg_marks[idx]:.2f}")
```

```python
print("\n" + "=" * 55)
print("           SCHOLARSHIP ELIGIBILITY")
print("=" * 55)
print("Criteria: Average >= 85 AND No subject below 70")
print("-" * 55)

#average >= 85
condition_avg = avg_marks >= 85

#no subject below 70 
condition_no_low = np.all(marks >= 70, axis=1)

#both conditions must be true
scholarship_mask = condition_avg & condition_no_low
scholarship_students = student_ids[scholarship_mask]

if len(scholarship_students) == 0:
    print("\n  No students are eligible for scholarship.")
else:
    print(f"\nEligible Students ({len(scholarship_students)} total):")
    for sid in scholarship_students:
        idx = sid - 1
        print(f"  Student {sid:>3}"
              f" | Average: {avg_marks[idx]:.2f}"
              f" | Total: {total_marks[idx]}")

print("\n" + "=" * 55)
print("         ANALYSIS COMPLETE")
print("=" * 55)
```
